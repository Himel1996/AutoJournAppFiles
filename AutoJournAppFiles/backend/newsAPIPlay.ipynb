```python
import requests
```

```python
def get_news_by_topic(topic, api_key, page_size=5):
    url = "https://newsapi.org/v2/everything"
    params = {
        'q': topic,
        'language': 'en',
        'sortBy': 'publishedAt',
        'pageSize': page_size,
        'apiKey': api_key
    }

    response = requests.get(url, params=params)
    data = response.json()

    if response.status_code != 200:
        print("Error:", data.get("message", "Unknown error"))
        return []

    articles = data.get("articles", [])
    for i, article in enumerate(articles):
        print(f"\nüîπ Article {i + 1}")
        print(f"Title: {article['title']}")
        print(f"Source: {article['source']['name']}")
        print(f"Published At: {article['publishedAt']}")
        print(f"URL: {article['url']}")
        print(f"Description: {article['description']}")

    return articles

# üß™ Example usage
if __name__ == "__main__":
    topic = "Trade war"
    api_key = "ce398b1406684c159597f7a51eea10cc"  # Replace this
    get_news_by_topic(topic, api_key)
```

```python
from newspaper import Article

def extract_full_text(article_url):
    try:
        article = Article(article_url)
        article.download()
        article.parse()
        return article.text
    except Exception as e:
        print(f"Error extracting article: {e}")
        return None

# Example usage
url = "https://www.globenewswire.com/news-release/2025/07/22/3119181/28124/en/July-2025-Tariffs-and-Trade-Wars-Strategic-Intelligence-Report-As-90-Day-Pause-Nears-End-Uncertainty-Looms-Over-U-S-Trade-Deals-and-Tariff-Levels.html"
full_text = extract_full_text(url)
print(full_text)
```

```python
from newspaper import Article
import requests

def get_full_articles(topic, api_key, max_articles=3):
    url = "https://newsapi.org/v2/everything"
    params = {
        'q': topic,
        'language': 'en',
        'sortBy': 'publishedAt',
        'pageSize': max_articles,
        'apiKey': api_key
    }

    response = requests.get(url, params=params).json()
    articles = response.get("articles", [])

    results = []
    for article in articles:
        print(f"üîπ {article['title']}")
        print(f"üîó {article['url']}")
        try:
            a = Article(article['url'])
            a.download()
            a.parse()
            print(f"üì∞ Full Text:\n{a.text[:500]}...\n")  # Truncated
            results.append((article['title'], a.text))
        except Exception as e:
            print(f"‚ö†Ô∏è Failed to extract: {e}")

    return results

# Replace with your own key
api_key = "ce398b1406684c159597f7a51eea10cc"
get_full_articles("Trade War", api_key)
```
